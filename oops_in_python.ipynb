```python
class Cat:
    def __init__(self,name,color,lives):
        self.name = name
        self.color = color
        self.lives = lives
    def meow(self):
        return f"Meow~ 🐾 I'm {self.name}, the {self.color} legend with {self.lives} lives... *naps*"

c = Cat("Mia","white",21)
print(c.__dict__)

print(c.meow())
```
